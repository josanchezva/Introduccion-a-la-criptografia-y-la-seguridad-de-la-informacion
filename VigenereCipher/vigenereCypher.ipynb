```python
#Vigenere Cypher

import numpy as np

# Split se usa para convertir el mensaje de string a lista
def split(word):
  return [char for char in word]
#Se usa para imprimir el mensaje final como un string desde una lista

def listToString(s): 
  str1 = "" 
  for ele in s: 
      str1 += ele  
  return str1 

def createVigenereMatrix(alphabet):
  alphabetValues = list(alphabet.values())
  matrix = []

  for i in range(len(alphabet)):
    for j in range(len(alphabet)):
      index = (i+j+26)%26
      value = alphabetValues[index]
      matrix.append(value)
  matrix = np.array(matrix)
  matrix= matrix.reshape((26,26))
  return matrix

if __name__ == '__main__':      
  print('0.Encriptar 1.Desencriptar')
  choice = input()
  #choice = '0'

  print('clave:')
  #key = 'CRYPTO'
  key = input()
  key = key.replace(' ', '')
  key = split(key)

  print('Parametro:')
  t = int(input())

  print('Mensaje:')
  #message = 'TO BE OR NOT TO BE THAT IS THE QUESTION'
  #message = 'KSMEH ZBBLK SMEMP OGAJX SEJCS FLZSY'
  #message = 'THERE IS A SECRET PASSAGE BEHIND THE PICTURE FRAME'
  message = input()
  message = message.replace(' ', '')
  message = split(message)



  #t = 3

  alphabet = {
    1:'A', 2:'B', 3:'C', 4:'D', 5:'E', 6:'F', 7:'G', 8:'H', 9:'I', 10:'J', 11:'K', 12:'L', 13:'M', 14:'N', 15:'O', 16:'P', 17:'Q', 18:'R', 19:'S', 20:'T', 21:'U', 22:'V', 23:'W', 24:'X', 25:'Y', 26:'Z'
  }

  alphabetKeys = list(alphabet.keys())
  alphabetValues = list(alphabet.values())

  matrix = createVigenereMatrix(alphabet)

  encriptedMessage = []
  decriptedMessage = []

  if(choice=='0'):
    for i in range(len(message)):
      keywordRowValue = key[i%len(key)]
      messageColumnValue = message[i]

      rowValue = alphabetValues.index(keywordRowValue)
      ColumnValue = alphabetValues.index(messageColumnValue)

      encriptedMessage.append(matrix[rowValue][ColumnValue])
      
      if((i+1)%t==0):
        encriptedMessage.append(' ')
      #print(messageColumnValue, keywordRowValue)
      #print(ColumnValue, rowValue)
    print('Mensaje encriptado: ',listToString(encriptedMessage))
  else: 
    for i in range(len(message)):
      keywordRowValue = key[i%len(key)]
      rowValue = alphabetValues.index(keywordRowValue)
      messageColumnValue = 0
      ColumnValue = 0

      for j in range(len(alphabetValues)):
        if(matrix[rowValue][j]==message[i]):
          messageColumnValue = matrix[0][j]
          ColumnValue = alphabetValues.index(messageColumnValue)

      decriptedMessage.append(matrix[0][ColumnValue])
      
      if((i+1)%t==0):
        decriptedMessage.append(' ')
      #print(messageColumnValue, keywordRowValue)
      #print(ColumnValue, rowValue)
    print('Mensaje decodificado: ',listToString(decriptedMessage))




  
```
