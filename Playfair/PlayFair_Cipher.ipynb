```python
import numpy as np

# Split se usa para convertir el mensaje y la clave de string a lista
def split(word):
    return [char for char in word]
#Se usa para imprimir el mensaje final como un string desde una lista

def listToString(s): 
    str1 = "" 
    for ele in s: 
        str1 += ele  
    return str1 

if __name__ == '__main__':

  print('0.Encriptar 1.Desencriptar')
  choice = input()
  alphabet=['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L', 'M', 'N', 'O', 'P', 'Q', 'R', 'S', 'T', 'U', 'V', 'W', 'X', 'Y', 'Z']

  print('Mensaje: ')
  mensaje=input()
  mensaje=mensaje.replace(' ', '') #se remueven los espacios en blancod el mensaje

  for i in range(0,len(mensaje)-1,2): #anade 'X' en parejas repetidos
    if(mensaje[i]==mensaje[i+1]):
      mensaje = mensaje[:i+1] + 'X' + mensaje[i+1:]

  mensaje = split(mensaje)

  if(len(mensaje)%2!=0):
    mensaje.append('X')
  print('Clave: ')

  clave=input()
  clave = split(clave)

  res = []
  #busca letras repetidas en la clave
  for i in clave:
    if i not in res:
        res.append(i)
  clave=res


  array=clave
  #Crea la matriz poniendo primero la clave y luego itera para que no se repitan los caracteres
  for i in range(len(alphabet)):
    exists= False
    for j in range(len(array)):
      if(array[j]==alphabet[i]):
        exists = True
    if (exists==False):
      array.append(alphabet[i])

  matrix5x5= np.array(array)
  matrix5x5= matrix5x5.reshape((5,5))




  #Encriptado
  if(choice=='0'):
    result=[]
    for i in range(0,len(mensaje)-1,2):
      #Toma parejas y busca sus posiciones en la matriz para comparar si tienen la misma fila columna o ninguna
      couple1=mensaje[i]
      couple2=mensaje[i+1]

      coupleInMatrixLeft=np.where(matrix5x5==couple1)
      coupleInMatrixRight=np.where(matrix5x5==couple2)


      if(coupleInMatrixLeft[1]==coupleInMatrixRight[1]):
        result.append((matrix5x5[(coupleInMatrixLeft[0][0]+1) %5] [coupleInMatrixLeft[1][0]]))
        result.append((matrix5x5[(coupleInMatrixRight[0][0]+1) %5] [coupleInMatrixRight[1][0]]))

      elif(coupleInMatrixLeft[0]==coupleInMatrixRight[0]):
        result.append((matrix5x5[coupleInMatrixLeft[0][0]] [(coupleInMatrixLeft[1][0]+1)%5]))
        result.append((matrix5x5[coupleInMatrixRight[0][0]] [(coupleInMatrixRight[1][0]+1)%5]))

      else:
        result.append((matrix5x5[coupleInMatrixLeft[0][0]] [coupleInMatrixRight[1][0]]))
        result.append((matrix5x5[coupleInMatrixRight[0][0]] [coupleInMatrixLeft[1][0]]))

    print('Playfair Matrix', matrix5x5)
    print('Mensaje final: ',listToString(result))
  #Desencriptado
  else:
    #Toma parejas y busca sus posiciones en la matriz para comparar si tienen la misma fila columna o ninguna
    result=[]
    for i in range(0,len(mensaje)-1,2):
      couple1=mensaje[i]
      couple2=mensaje[i+1]

      coupleInMatrixLeft=np.where(matrix5x5==couple1)
      coupleInMatrixRight=np.where(matrix5x5==couple2)


      if(coupleInMatrixLeft[1]==coupleInMatrixRight[1]):
        result.append((matrix5x5[(coupleInMatrixLeft[0][0]-1) %5] [coupleInMatrixLeft[1][0]]))
        result.append((matrix5x5[(coupleInMatrixRight[0][0]-1) %5] [coupleInMatrixRight[1][0]]))

      elif(coupleInMatrixLeft[0]==coupleInMatrixRight[0]):
        result.append((matrix5x5[coupleInMatrixLeft[0][0]] [(coupleInMatrixLeft[1][0]-1)%5]))
        result.append((matrix5x5[coupleInMatrixRight[0][0]] [(coupleInMatrixRight[1][0]-1)%5]))

      else:
        result.append((matrix5x5[coupleInMatrixLeft[0][0]] [coupleInMatrixRight[1][0]]))
        result.append((matrix5x5[coupleInMatrixRight[0][0]] [coupleInMatrixLeft[1][0]]))

    print('Playfair Matrix', matrix5x5)
    print('Mensaje final: ',listToString(result))
```
