```python
#Hill Cipher
import numpy as np
import sympy as sp
import sys

# Split se usa para convertir el mensaje de string a lista
def split(word):
  return [char for char in word]
#Se usa para imprimir el mensaje final como un string desde una lista

def listToString(s): 
  str1 = "" 
  for ele in s: 
      str1 += ele  
  return str1

def gcdExtended(a, b): 
    # Base Case 
    if a == 0 :  
        return b,0,1
             
    gcd,x1,y1 = gcdExtended(b%a, a) 
     
    # Update x and y using results of recursive 
    # call 
    x = y1 - (b//a) * x1 
    y = x1 
     
    return gcd,x,y

if __name__ == '__main__':

  alphabet = {
    0:'A', 1:'B', 2:'C', 3:'D', 4:'E', 5:'F', 6:'G', 7:'H', 8:'I', 9:'J', 10:'K', 11:'L', 12:'M', 13:'N', 14:'O', 15:'P', 16:'Q', 17:'R', 18:'S', 19:'T', 20:'U', 21:'V', 22:'W', 23:'X', 24:'Y', 25:'Z'
  }

  alphabetKeys = list(alphabet.keys())
  alphabetValues = list(alphabet.values())

  print('0.Encriptar 1.Desencriptar')
  choice = input()
  #choice = '1'

  print('clave (M_2x2) 4 Numeros separados por espacios : (EJ: 11 8 3 7)')
  #key = '11 8 3 7'
  key = input()
  key = key.split()
  key = np.array(key)
  key = key.astype(int)
  key= key.reshape((2,2))
  detKey = round(np.linalg.det(key))
  gcd, x, y = gcdExtended(detKey,len(alphabet))

  print(key)

  if(detKey== 0):
    print('La matriz no es invertible')
    sys.exit()
  elif (gcd != 1):
    print('El gcd del determinante de la llave y el tama√±o del alfabeto es distinto de 1')
    sys.exit()

  

  print('Mensaje:')
  #message = 'VFDNRQNLPJKM'
  message = input()
  message = message.replace(' ', '')
  message = split(message)

  

  if(choice=='0'):
    encriptedMessage = []
    for i in range(0,len(message)-1,2):
      couple1 = message[i]
      couple2 = message[i+1]

      value1 = alphabetValues.index(couple1)
      value2 = alphabetValues.index(couple2)

      aux = []
      aux.append(value1)
      aux.append(value2)

      res = np.matmul(aux,key)%26

      encriptedMessage.append(res[0])
      encriptedMessage.append(res[1])


    for i in range(len(encriptedMessage)):
      encriptedMessage[i] = alphabet.get(encriptedMessage[i])
    
    print('Mensaje encriptado: ',listToString(encriptedMessage))
  else:
    decriptedMessage = []
    detKey = round(np.linalg.det(key))
    adjointKey = sp.Matrix(key).adjugate()

    gcd,x,y = gcdExtended(detKey,len(alphabet))

    inverseMatrix = (x*adjointKey)%26
    inverseMatrix = np.array(inverseMatrix)
    

    for i in range(0,len(message)-1,2):
      couple1 = message[i]
      couple2 = message[i+1]

      value1 = alphabetValues.index(couple1)
      value2 = alphabetValues.index(couple2)

      aux = []
      aux.append(value1)
      aux.append(value2) 
      
      res = np.matmul(aux,inverseMatrix)%26

      decriptedMessage.append(res[0])
      decriptedMessage.append(res[1])


    for i in range(len(decriptedMessage)):
      decriptedMessage[i] = alphabet.get(decriptedMessage[i])
    print('Mensaje desencriptado:' ,listToString(decriptedMessage))
    


```
